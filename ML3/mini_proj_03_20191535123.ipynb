```python
import numpy as np
import pandas as pd


input_file = "kc_house_data.csv"
df = pd.read_csv(input_file)
```

```python
from sklearn.utils import shuffle

df = shuffle(df)
```

```python
df.head()
```

```python
df = df.drop(["date"], axis=1)
```

```python
df.info()
```

## 특성 추가"

```python
df['sqft_living*grade'] = df['sqft_living'] * df['grade']
df['sqft_living*sqft_above'] = df['sqft_living'] * df['sqft_above']
df['sqft_living*sqft_living15'] = df['sqft_living'] * df['sqft_living15']
df['sqft_living_sum_grade'] = df['sqft_living'] + df['grade']
df['sqft_living_sum_sqft_above'] = df['sqft_living'] + df['sqft_above']
df['sqft_living_sum_sqft_living15'] = df['sqft_living'] + df['sqft_living15']
df['sqft_living_sum_sqft_bathrooms'] = df['sqft_living'] + df['bathrooms']
df['sqft_living_sum_sqft_view'] = df['sqft_living'] + df['view']
df['total_rooms'] = df['bedrooms'] + df['bathrooms']
df['living_to_lot_ratio'] = df['sqft_living'] / df['sqft_lot']
df['total_area'] = df['sqft_living'] + df['sqft_lot'] + df['sqft_above'] + df['sqft_basement']
df['renovated'] = df['yr_renovated'].apply(lambda x: 0 if x == 0 else 1)
df['high_priced_area'] = df['zipcode'].apply(lambda x: 1 if x in [98001, 98002, 98003] else 0)
df['view_quality'] = df['waterfront'] * df['view']
df['total_rooms'] = df['bedrooms'] + df['bathrooms']
df['overall_condition'] = df['condition'] + df['grade']

# 결과 확인
df.head()
```

```python
corr_matrix = df.corr()
```

```python
corr_matrix["price"].sort_values(ascending=False)
```

```python
housing_prepared = df.drop(["price"], axis=1)
housing_labels = df["price"].copy()
```

```python
housing_prepared.describe()
```

## 이상치 대체 하기

```python
def replace_outliers_with_median(df):
    variables = list(df.columns.values)
    for variable in variables:
        median_value = df[variable].median()
        std_value = df[variable].std()
        outliers = (df[variable] - median_value).abs() > 3 * std_value
        df.loc[outliers, variable] = median_value
    return df
```

```python
housing_prepared_outliers = replace_outliers_with_median(housing_prepared)
```

```python
housing_prepared_outliers.describe()
```

```python
def display_scores(scores):
    print('Scores:', scores)
    print('Mean:', scores.mean())
    print('Standard deviation:', scores.std())
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from lightgbm import LGBMRegressor

lgbm = LGBMRegressor(    max_depth=10,  # Adjust max_depth based on your preference
    min_data_in_leaf=29,
    feature_fraction=1.0,
    n_estimators=150,
    bagging_fraction=0.1,
    min_gain_to_split=1,
    learning_rate=0.1,
    num_leaves=84,  # Set num_leaves based on your preference
    random_state=50,
    force_col_wise=True
)
score = cross_val_score(lgbm, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv = 10)
rmse_scores = np.sqrt(-score)
```

```python
display_scores(rmse_scores)
```
